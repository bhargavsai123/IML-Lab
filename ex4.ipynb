```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("ultimate_student_productivity_dataset_5000.csv")
print(data.head())
```

```python
X = data.select_dtypes(include=np.number)
X = X.drop(columns=['exam_score','student_id'])
y = data['exam_score']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
model = LinearRegression()

X_train, X_test, y_train, y_test = 
train_test_split(X, y, test_size=0.3, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Before Feature Selection ")
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
correlation = X.corrwith(y)
print(correlation.sort_values(ascending = False))
X_corr = correlation[correlation > 0.5].index
print("Correlation Selected Features : ",X_corr)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X[X_corr], y, test_size=0.3, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("After Correlation ")
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
from sklearn.feature_selection import RFE

rfe = RFE(model,n_features_to_select = 4)
rfe.fit(X,y)

selected_features_rfe = X.columns[rfe.support_]
print("Selected Features (RFE) : ",selected_features_rfe)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X[selected_features_rfe], y, test_size=0.3, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("After RFE ")
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.1)
lasso.fit(X,y)
lasso_coeffs = pd.Series(lasso.coef_, index=X.columns)
print(lasso_coeffs.sort_values(ascending=False))
X_lasso = lasso_coeffs[lasso_coeffs > 0].index
print("Lasso Selected Features : ",X_lasso)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X[X_lasso], y, test_size=0.3, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("After Embedded (Lasso) ")
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)
print(X_standardized[:5])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_standardized, y, test_size=0.3, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("After Standardization ")
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)
print(X_normalized[:5])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.3, random_state=42)
model2 = LinearRegression()
model2.fit(X_train, y_train)
y_pred = model2.predict(X_test)
print("After Normalization ")
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```
