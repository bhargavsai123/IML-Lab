```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data.csv')
print(data.head())
print("Shape : ",data.shape)
```

```python
data = data.drop(columns=['Rating','Reviews','SSD'])
data_missing = data.copy()
print("Missing Values Before Handling: ")
print(data_missing.isnull().sum())
```

```python
data_missing.fillna({'Core':'--','Actual Price':data_missing['Discounted Price'],'Saving':'0% OFF'},inplace=True)
print("Missing values after handling: ")
print(data_missing.isnull().sum())
```

```python
from scipy.stats import zscore

numeric_data = data_missing.select_dtypes(include=np.number)
z_scores = np.abs(zscore(numeric_data))
threshold = 3
data_cleaned = data_missing[(z_scores < threshold).all(axis=1)]
print("Shape after outlier treatment:", data_cleaned.shape)
```

```python
data_cleaned.to_csv('cleaned_data.csv', index=False)
```
